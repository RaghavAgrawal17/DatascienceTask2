```python
import urllib.request
import re
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/RaghavAgrawal17/practice-Datasets/main/task.txt"
file = urllib.request.urlopen(url)

content = ''

for line in file:
  line = " ".join(line.decode("utf-8").split())
  if line:
    content += line
    content += ' '

print(content)
```

```python
#Converting content to lowercase
content = content.lower()

#Removing all the special characters from the text
pattern = r'[^0-9a-zA-Z]+'
content = re.sub(pattern, ' ', content)

#Removing all the numbers from the text making words
pattern = r'[0-9]'
content = re.sub(pattern, '', content)

content = ' '.join(content.split())

"""
#Removing stop words 
stop_words = set(stopwords.words('english'))
word_tokens = content.split(' ')
filtered_sentence = []

for w in word_tokens:
  if w not in stop_words:
    filtered_sentence.append(w)

print(filtered_sentence)
"""

filtered_sentence = content.split()
print(filtered_sentence)
```

```python
def density_oneword(words):
  unique_words = set(words)
  n = len(words)
  keyword = []
  density_count = []
  density_percentage = []

  for w in unique_words:
    result = words.count(w)
    percentage = str(round(100*result/n,2))+'%'
    keyword.append(w)
    density_count.append(result)
    density_percentage.append(percentage)

  density_percentage, density_count, keyword = zip(*sorted(zip(density_percentage, density_count, keyword),reverse= True))

  return density_percentage, density_count, keyword
```

```python
def density_twoword(words):
  twowordlist = []

  for x,y in zip(words[::1],words[1::1]):
    twowords = x+' '+y
    twowordlist.append(twowords)
  
  # Same as before but now the new list of 2 joined words is used instead
  uniquetwowords = set(twowordlist)
  n = len(twowordlist)

  keyword = []
  density_count = []
  density_percentage = []

  for word in uniquetwowords:
    result = twowordlist.count(word)
    percentage = str(round(100*result/n,2))+'%'

    keyword.append(word)
    density_count.append(result)
    density_percentage.append(percentage)

  density_percentage, density_count, keyword = zip(*sorted(zip(density_percentage, density_count, keyword),reverse=True))

  return density_percentage, density_count, keyword
```

```python
def density_threeword(words):
  threewordlist = []

  for x,y,z in zip(words[::1],words[1::1],words[2::1]):
    threewords = x+' '+y+' '+z
    threewordlist.append(threewords)

  uniquethreewords = set(threewordlist)
  n = len(threewordlist)

  keyword = []
  density_count = []
  density_percentage = []

  for word in uniquethreewords:
    result = threewordlist.count(word)
    percentage = str(round(100*result/n,2))+'%'

    keyword.append(word)
    density_count.append(result)
    density_percentage.append(percentage)
    
	
  density_percentage, density_count, keyword = zip(*sorted(zip(density_percentage, density_count, keyword),reverse = True))

  return density_percentage, density_count, keyword
```

```python
keyword_density_output = []
```

```python
list0 = []
a, b, c = density_oneword(filtered_sentence)
list0.append(list(a))
list0.append(list(b))
list0.append(list(c))
```

```python
list1 = []
a, b, c = density_twoword(filtered_sentence)
list1.append(list(a))
list1.append(list(b))
list1.append(list(c))
```

```python
list2 = []
a, b, c = density_threeword(filtered_sentence)
list2.append(list(a))
list2.append(list(b))
list2.append(list(c))
```

```python
#Required Output

keyword_density_output.append(list(list0))
keyword_density_output.append(list(list1))
keyword_density_output.append(list(list2))

len(keyword_density_output)
```

```python
data = np.array(keyword_density_output[0]).T
df = pd.DataFrame(data, columns=['Density', 'Count', 'Word'])
df
```
